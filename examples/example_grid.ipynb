# Grid graph using networkx

```python
import networkx as nx
import popy


class MyAgent(popy.Agent):
    pass


class Grid(popy.MagicLocation):
    nxgraph = nx.grid_2d_graph(4, 4)


model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)

for _ in range(16):
    popy.Agent(model)

creator.create_locations(location_classes=[Grid])
```
