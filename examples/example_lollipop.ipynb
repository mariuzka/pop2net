# Lollipop graph

## Example I - Simple Lollipop using networkx

```python
import networkx as nx
import popy

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)


class Lollipop(popy.MagicLocation):
    n_agents = 7
    nxgraph = nx.lollipop_graph(5, 3)


for _ in range(20):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Lollipop,
    ]
)
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```

## Example II - A Lollipop flower

```python
model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)


class Lollipop(popy.MagicLocation):
    n_locations = 5
    nxgraph = nx.lollipop_graph(5, 3)
    only_exact_n_agents = True


class Center(popy.MagicLocation):
    def filter(self, agent):
        return agent.Lollipop_tail or not agent.Lollipop

    def bridge(self, agent):
        return agent.Lollipop_assigned


for _ in range(41):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Lollipop,
        Center,
    ]
)
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```
