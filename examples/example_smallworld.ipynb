# Small world using networkx

```python
import networkx as nx
import popy


class MyAgent(popy.Agent):
    pass


class SmallWorld(popy.MagicLocation):
    nxgraph = nx.watts_strogatz_graph(n=50, k=4, p=0.05)


model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)

for _ in range(50):
    popy.Agent(model)

creator.create_locations(location_classes=[SmallWorld])
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```
