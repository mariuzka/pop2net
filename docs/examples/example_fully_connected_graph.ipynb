# Complete graph

## Example I: Simple complete graph using p2n only

```python
import pop2net as p2n


class FullyConnected(p2n.MagicLocation):
    pass


model = p2n.Model()
creator = p2n.Creator(model)
inspector = p2n.NetworkInspector(model)

for _ in range(5):
    p2n.Agent(model)

creator.create_locations(location_classes=[FullyConnected])
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```

## Example II: Complete graph with one location for each bilateral relation using networkx

```python
import networkx as nx


class FullyConnected(p2n.MagicLocation):
    nxgraph = nx.complete_graph(n=5)


model = p2n.Model()
creator = p2n.Creator(model)
inspector = p2n.NetworkInspector(model)

for _ in range(5):
    p2n.Agent(model)

creator.create_locations(location_classes=[FullyConnected])
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```
