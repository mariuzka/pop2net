# Fully connected graph

```python
import popy


class MyAgent(popy.Agent):
    pass


class FullyConnected(popy.MagicLocation):
    pass


model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)

for _ in range(10):
    popy.Agent(model)

creator.create_locations(location_classes=[FullyConnected])
```

```python
inspector.plot_bipartite_network()
```
