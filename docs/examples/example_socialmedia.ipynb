```python
import pop2net as p2n

env = p2n.Environment()
creator = p2n.Creator(env=env)
inspector = p2n.NetworkInspector(env=env)

class DirectedLocation(p2n.Location):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.origins = []
        self.targets = []
    
    def project_weights(self, actor1, actor2):
        if actor1.id_p2n in self.origins and actor2.id_p2n in self.origins:
            return 0
        elif actor1.id_p2n in self.targets and actor2.id_p2n in self.targets:
            return 0
        else:
            return 1

class User(p2n.Actor):
    pass

class Follow(DirectedLocation):
    pass


for i in range(10):
    user = User()
    env.add_actor(user)

follow0 = Follow()
env.add_location(follow0)

follow0.origins.append(env.actors[0].id_p2n)
follow0.add_actor(env.actors[0])

follow0.targets.extend([actor.id_p2n for actor in env.actors[1:]])
follow0.add_actors(env.actors[1:])

inspector.plot_networks(include_0_weights=False)
```
