# Caveman graph

## Example I - Unconnected caves

```python
import popy

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)


class Cave(popy.MagicLocation):
    n_agents = 5


for _ in range(25):
    popy.Agent(model)

creator.create_locations(location_classes=[Cave])
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```

## Example II - Connected grid caves

```python
import networkx as nx

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)


class Cave(popy.MagicLocation):
    n_agents = 9
    nxgraph = nx.grid_2d_graph(3, 3)
    only_exact_n_agents = True


class Bridge(popy.MagicLocation):
    nxgraph = nx.cycle_graph(5)

    def filter(self, agent):
        return agent.Cave_head

    def bridge(self, agent):
        return agent.Cave


for _ in range(45):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Cave,
        Bridge,
    ]
)
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```
