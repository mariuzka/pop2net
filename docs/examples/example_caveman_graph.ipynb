# Caveman graph

## Example I

```python
import popy

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)

class Cave(popy.MagicLocation):
    n_agents = 5

for _ in range(25):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Cave
        ]
    )
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```

## Example II

```python
import popy

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)

class Cave(popy.MagicLocation):
    n_locations = 3

for _ in range(25):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Cave,
        ]
    )
```
