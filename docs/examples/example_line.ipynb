# Line graph using networkx

## Example I

```python
import networkx as nx
import popy

model = popy.Model()
creator = popy.Creator(model)
inspector = popy.NetworkInspector(model)


class Line(popy.MagicLocation):
    nxgraph = nx.path_graph(10)


for _ in range(10):
    popy.Agent(model)

creator.create_locations(
    location_classes=[
        Line,
    ]
)
```

```python
inspector.plot_bipartite_network()
```

```python
inspector.plot_agent_network()
```
