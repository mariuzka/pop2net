```python
import popy
import networkx as nx
from bokehgraph import BokehGraph
```

```python
model = popy.Model()
```

```python
agent = popy.Agent(model)
```

```python
location = popy.Location(model)
```

```python
graph = model.g.copy()
for i in graph:
    graph.nodes[i]["id"] = getattr(graph.nodes[i]["_obj"], "id")
    del graph.nodes[i]["_obj"]

graph_layout = nx.drawing.spring_layout(graph)
plot = BokehGraph(graph, width=500, height=500, hover_edges=True)
plot.layout(layout=graph_layout)
plot.draw(
)
```

```python
location.add_agent(agent)
```
