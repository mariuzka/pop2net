```python
import agentpy as ap
import mesa
import pop2net as p2n
```

## No framework

```python
class MyLocationDesigner(p2n.LocationDesigner):
    pass


env = p2n.Environment()
creator = p2n.Creator(env=env)
inspector = p2n.NetworkInspector(env=env)

creator.create_actors(n=10)
creator.create_locations(location_designers=[MyLocationDesigner])
inspector.plot_bipartite_network()
```

## Pop2net + Agentpy

```python
class MyLocationDesigner(p2n.LocationDesigner):
    pass


model = ap.Model()
env = p2n.Environment(model=model, framework="agentpy")
creator = p2n.Creator(env=env)
inspector = p2n.NetworkInspector(env=env)

creator.create_actors(n=10)
creator.create_locations(location_designers=[MyLocationDesigner])
inspector.plot_bipartite_network()
```
